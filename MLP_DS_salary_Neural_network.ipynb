```python
import pandas as pd
import tensorflow as tf
import os
import datetime
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense, BatchNormalization
from tensorflow.keras.models import Sequential
```

```python
data = pd.read_csv('ds_salaries.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
y = data['salary_in_usd']
```

```python
X = data.drop('salary_in_usd', axis=1)
```

```python
data.info()
```

```python
# Identyfikacja kolumn typu object
object_columns = X.select_dtypes(include='object').columns
print(f"\nColumns of type object: {object_columns}")
```

```python
# Inicjalizacja LabelEncoder
label_encoders = {}

# Kodowanie wartości kategorycznych
for col in object_columns:
    if X[col].dtype == 'object':
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col])
        label_encoders[col] = le
```

```python
logdir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = tf.keras.callbacks.TensorBoard(logdir)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

### Model 1 - with sigmoid activation

```python
model = Sequential()
model.add(Dense(20, input_dim=X_train.shape[1], activation='sigmoid'))
```

```python
# Dodanie 8 warstw Gęstych z 20 neuronami każda
for _ in range(7):
    model.add(Dense(20, activation='sigmoid'))

# Dodanie warstwy wyjściowej
model.add(Dense(1, activation='linear'))

# Kompilowanie modelu
model.compile(optimizer='adam', loss='mean_squared_error')

# Trenowanie modelu
model.fit(X_train, y_train, epochs=100, batch_size=5, verbose=1)
```

```python
# Ocena modelu na danych testowych
loss = model.evaluate(X_test, y_test)
print(f'Strata modelu na danych testowych: {loss:.2f}')
```

It looks like model is not working well on testing data, let's try use relu as activaction function.  

### Model 2 - with relu activation

```python
model2 = Sequential()
model2.add(Dense(20, input_dim=X_train.shape[1], activation='relu'))
```

```python
# Dodanie 8 warstw Gęstych z 20 neuronami każda
for _ in range(7):
    model2.add(Dense(20, activation='relu'))

# Dodanie warstwy wyjściowej
model2.add(Dense(1, activation='linear'))

# Kompilowanie modelu
model2.compile(optimizer='adam', loss='mean_squared_error')

# Trenowanie modelu
model2.fit(X_train, y_train, epochs=100, batch_size=5, verbose=1)
```

```python
# Ocena modelu na danych testowych
loss_2 = model2.evaluate(X_test, y_test)
print(f'Strata modelu na danych testowych: {loss_2:.2f}')
```

### Model 3

```python
model3 = Sequential()
model3.add(Dense(20, input_dim=X_train.shape[1], activation='relu'))
model3.add(BatchNormalization())
```

```python
# Dodanie 8 warstw Gęstych z 20 neuronami każda
for _ in range(7):
    model3.add(Dense(20, activation='relu'))

# Dodanie warstwy wyjściowej
model3.add(Dense(1, activation='linear'))

# Kompilowanie modelu
model3.compile(optimizer='adam', loss='mean_squared_error')

# Trenowanie modelu
model3.fit(X_train, y_train, epochs=100, batch_size=5, verbose=1)
```

```python
loss3 = model3.evaluate(X_test, y_test)
print(f'Strata modelu na danych testowych: {loss3:.2f}')
```
